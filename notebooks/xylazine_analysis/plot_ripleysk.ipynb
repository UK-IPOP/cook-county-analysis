```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio

pio.renderers.default = "notebook"
```

```python
df1 = pd.read_csv("./alcohol_compared_pattern.csv")
df2 = pd.read_csv("./fentanyl_compared_pattern.csv")
df1.head()
```

```python
df1['control'] = 'alcohol'
df2['control'] = 'fentanyl'
df1.head()
```

```python
df = pd.concat([df1, df2], ignore_index=True)
print(df.shape)
df.head()
```

```python
print("xylazine X alcohol VS xylazine X fentanyl")
(df1['obs'] - df2['obs']).describe()['mean']
```

```python
fig = px.line(df, x='r', y='obs', color='control')
fig.add_scatter(x=df['r'], y=df['theo'], name='theoretical')
fig.add_scatter(x=df1['r'], y=df1['lo'], name='alcohol_lo_CI')
fig.add_scatter(x=df1['r'], y=df1['hi'], name='alcohol_hi_CI')
fig.add_scatter(x=df2['r'], y=df2['lo'], name='fentanyl_lo_CI')
fig.add_scatter(x=df2['r'], y=df2['hi'], name='fentanyl_hi_CI')
fig.add_scatter(x=df1['r'], y=(df1['obs'] - df2['obs']), name='alc-fent diff')
fig.update_layout(title="Ripley 's K Differences, Baseline Case is Xylazine", xaxis_title='r', yaxis_title='observedK Diff')
fig.write_html("./ripleysk_plot.html")
fig.show()
```
